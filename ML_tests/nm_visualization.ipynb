```python
import requests
import json
import matplotlib.pyplot as plt
x1 = []
y1 = []
z1 = []
x2 = []
y2 = []
z2 = []
```

```python
x1.clear()
y1.clear()
z1.clear()
x2.clear()
y2.clear()
z2.clear()
f = requests.post(url="http://localhost:1337/rk/2", json={
  "a": 0,
  "b": 3.14,
  "n": 100
})

g = requests.post(url="http://localhost:1337/rk/4", json={
  "a": 0,
  "b": 3.14,
  "n": 100
})

jfile1 = json.loads(f.text)
for item in jfile1:
    x1.append(item['x'])
    y1.append(item['y'])
    z1.append(item['z'])
    
jfile2 = json.loads(g.text)
for item in jfile2:
    x2.append(item['x'])
    y2.append(item['y'])
    z2.append(item['z'])


plt.subplot(121)
plt.title("rk2")
plt.plot(x1, y1)
plt.plot(x1, z1)
plt.xlabel("x")
plt.legend(["y", "z"], loc=0, frameon=True)
plt.subplot(122)
plt.title("rk4")
plt.plot(x2, y2)
plt.plot(x2, z2)
plt.xlabel("x")
plt.legend(["y", "z"], loc=0, frameon=True)
plt.show()
# print('y1: ')
# print(y1)
# print('z1: ')
# print(z1)
```
